```python
#find R when R1 = 50 and R2 = 75 with the formula provided: (1/R) = (1/R1) + (1/R2)
R1 = 50
R2 = 75

inverse_R = (1/R1) + (1/R2) #inverse_R = (1/R)
R = (1/inverse_R)
print(f"R is {R} when R1 is 50 and R2 is 75.")

#check answer
inverse_R = (1/30)
inverse_R1R2 = (1/50) + (1/75)
if inverse_R == inverse_R1R2:
    print("This answer is confirmed using substitution.")
else:
    print("The answer achieved is incorrect.")
```

```python
#find rate_R which is dR/dt
import sympy as sym #module for calculus functions

rate_R1 = 1 #dR1/dt
rate_R2 = 1.5

R_Sym = sym.Symbol('R') #initialize calculus variables
R1_Sym = sym.Symbol('R1')
R2_Sym = sym.Symbol('R2')

first_derivative_R = sym.diff((1/R_Sym)) #find first deriviative of R
first_derivative_R1 = sym.diff((1/R1_Sym))
first_derivative_R2 = sym.diff((1/R2_Sym))

first_derivative_R = -1 * first_derivative_R #multiple by -1 to remove negative sign
first_derivative_R1 = -1 * first_derivative_R1
first_derivative_R2 = -1 * first_derivative_R2

rate_R_equation = ((first_derivative_R1 * rate_R1) + (first_derivative_R2 * rate_R2)) / first_derivative_R
print(rate_R_equation)
```
